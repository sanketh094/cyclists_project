```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
```

```python
# importing and combining Quarter 1 data
df = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q1\202104-divvy-tripdata.csv")
d_f = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q1\202103-divvy-tripdata.csv")
f_d = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q1\202102-divvy-tripdata.csv")
m_d = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q1\202101-divvy-tripdata.csv")
data_Q1 = pd.concat([df,d_f,f_d,m_d],axis = 0)


```

```python
# importing and combining Quarterly  data
n_a = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q2\202105-divvy-tripdata.csv")
n_b = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q2\202106-divvy-tripdata.csv")
n_c = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q2\202107-divvy-tripdata.csv")
n_d = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q2\202108-divvy-tripdata.csv")
data_Q2 = pd.concat([n_a,n_b,n_c,n_d],axis = 0)

```

```python
# importing and saving into Quarterly 3 data
n_e = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q3\202109-divvy-tripdata.csv")
n_f = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q3\202110-divvy-tripdata.csv")
n_g = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q3\202111-divvy-tripdata.csv")
n_h = pd.read_csv(r"C:\Users\91935\Downloads\Cyclists data\Q3\202112-divvy-tripdata.csv")
data_Q3 = pd.concat([n_e,n_f,n_g,n_h],axis = 0)
data_Q3.head(50)
```

```python
# combining all data into one single data and add into a data frame
cyclists_data = pd.concat([data_Q1,data_Q2,data_Q3],axis = 0)
cyclists_data = pd.DataFrame(data = cyclists_data)
cyclists_data.head()
```

```python
# converting the started date and ended date to date formats
cyclists_data["started_at"] = pd.to_datetime(cyclists_data["started_at"])
cyclists_data["ended_at"] = pd.to_datetime(cyclists_data["ended_at"])
cyclists_data.head()
```

```python
# cleaning the data and removing the null or empty values
cyclists_data = cyclists_data.dropna(axis = "columns")
```

```python
# converting the started at and ended at to month date
cyclists_data["month"] = cyclists_data["started_at"].dt.strftime("%Y-%m")
cyclists_data["week_name"] = cyclists_data["started_at"].dt.day_name()
cyclists_data["month_name"] = cyclists_data["started_at"].dt.month_name()
cyclists_data["Quarter"] = cyclists_data["started_at"].dt.to_period("Q").dt.strftime("%Y-Q%q")
cyclists_data.head()
```

```python
# calucalting trip duration for the cyclists data
# trip_duration in minutes
cyclists_data["trip_duration_in_minutes"] = (cyclists_data["ended_at"] - cyclists_data["started_at"])
cyclists_data.head()
```

```python
# deleting the unnecessary columns
columns_to_delete = ["trip_duration",'trip_duration_weeks',"ride_id"]
cyclists_data.drop(columns = columns_to_delete,axis = 1,inplace = True)
```

```python
# converting the trip duration to numveric
cyclists_data["trip_duration_in_minutes"] = pd.to_numeric(cyclists_data["trip_duration_in_minutes"])
```

```python
# grouping the members on   casusal members
casual_trips = cyclists_data[cyclists_data["member_casual"] == "casual"]
casual_trips.head()
```

```python
# grouping the members on subscribed members
member_trips = cyclists_data[cyclists_data["member_casual"] == "member"]
member_trips.head()
```

```python
# grouping the data by weeks and their duration
grouped_data_casual = casual_trips.groupby(["trip_duration_in_minutes","rideable_type","week_name","member_casual"]) ["trip_duration_in_minutes"].agg(["mean","sum"]).reset_index()
grouped_data_casual.head()
```

```python
# grouping the data for the subscribed members
grouped_data_member = member_trips.groupby(["rideable_type","week_name","member_casual"]) ["trip_duration_in_minutes"].agg(["mean","sum"]).reset_index()
grouped_data_member.head()
```

```python
# converting the trip duration to numveric
cyclists_data["trip_duration_in_minutes"] = pd.to_numeric(cyclists_data["trip_duration_in_minutes"])
```

```python
# plotting these data to get quick visualizations for members
plt.figure(figsize=(10,6))
sns.barplot(x= "week_name", y="mean",hue = "rideable_type",data = grouped_data_member,ci = None)
plt.title("Average Trip Duration by Members, Rideable Type, and Week")
plt.xlabel("Week Name")
plt.ylabel("Average Trip Duration(minutes)")
plt.legend(title = "Rideable Type",bbox_to_anchor = (1.05,1),loc = "upper left")
plt.show()
```

```python
# plotting these data for casual members
plt.figure(figsize=(10,6))
sns.barplot(x= "week_name", y="mean",hue = "rideable_type",data = grouped_data_casual,ci = None)
plt.title("Average Trip Duration by Members, Rideable Type, and Week")
plt.xlabel("Week Name")
plt.ylabel("Average Trip Duration(minutes)")
plt.legend(title = "Rideable Type",bbox_to_anchor = (1.05,1),loc = "upper left")
plt.show()
```

```python
# creating plot for the all members and their rideable preference
grouped_data = cyclists_data.groupby(["member_casual","rideable_type","month_name"]) ["trip_duration_in_minutes"].agg(["mean","sum","count"]).reset_index()
grouped_data
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x= "month_name",y = "count",hue = "member_casual",data = grouped_data,ci = None)
plt.xlabel("Month Name")
plt.ylabel("Avg Trip Duration")
plt.title("Average Trip Duration By Rideable Type")
plt.xticks(rotation = 45)
plt.legend(title = "members",bbox_to_anchor = (1.05,1),loc= "upper left")
plt.show()
```
